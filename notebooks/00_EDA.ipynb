## 1. Análise Exploratória de Dados (EDA)
 Nesta etapa, exploramos a estrutura dos arquivos da CVM para identificar inconsistências, valores nulos e entender a distribuição das contas contábeis. Esta análise orienta as decisões de limpeza que virão a seguir.

```python
import pandas as pd
import numpy as np
```

### 1.1 Carregamento dos dados

```python
df_dre = pd.read_csv(
    "dfp_cia_aberta_DRE_con_2024.csv",
    sep=";",
    encoding="ISO-8859-1"
)

df_bpa = pd.read_csv(
    "dfp_cia_aberta_BPA_con_2024.csv",
    sep=";",
    encoding="ISO-8859-1"
)

df_bpp = pd.read_csv(
    "dfp_cia_aberta_BPP_con_2024.csv",
    sep=";",
    encoding="ISO-8859-1"
)
```

### 1.2 Inspeção estrutural dos datasets

```python
df_dre.head()
```

```python
df_bpa.head()
```

```python
df_bpp.head()
```

### 1.3 Qualidade dos dados

```python
datasets = {
    "DRE": df_dre,
    "BPA": df_bpa,
    "BPP": df_bpp
}

for nome, df in datasets.items():

    print(f"\n===== {nome} =====")

    # Visão inicial
    print("\nShape:")
    print(df.shape)

    print("\nInfo:")
    df.info()

    # Estatísticas básicas
    print("\nDescribe:")
    print(df.describe())

    # Valores nulos
    print("\nValores nulos:")
    print(df.isna().sum())

    # Empresas únicas
    print("\nEmpresas únicas (CNPJ):")
    print(df['CNPJ_CIA'].nunique())

    # Escala monetária
    print("\nEscala de moeda:")
    print(df['ESCALA_MOEDA'].unique())

    # Duplicados
    print("\nRegistros duplicados:")
    print(df.duplicated().sum())

    # Contas contábeis mais frequentes (se existir)
    if 'DS_CONTA' in df.columns:
        print("\nContas contábeis mais frequentes:")
        print(df['DS_CONTA'].value_counts().head(10))
```

### 1.4 Identificação das contas contábeis

```python
df_dre[['CD_CONTA', 'DS_CONTA']].drop_duplicates().sort_values('CD_CONTA')
```

```python
df_dre[df_dre['DS_CONTA'].str.contains("receita", case=False, na=False)][['CD_CONTA','DS_CONTA']].drop_duplicates()
```

```python
df_dre[df_dre['DS_CONTA'].str.contains("resultado", case=False, na=False)][['CD_CONTA','DS_CONTA']].drop_duplicates()
```

```python
df_bpa[['CD_CONTA','DS_CONTA']].drop_duplicates().sort_values('CD_CONTA')
```

```python
df_bpp[['CD_CONTA','DS_CONTA']].drop_duplicates().sort_values('CD_CONTA')
```

```python
datasets = {
    "DRE": df_dre,
    "BPA": df_bpa,
    "BPP": df_bpp
}

for nome, df in datasets.items():

    print(f"\n===== {nome} =====")

    # Distribuição da escala monetária
    print("\nDistribuição da escala monetária:")
    print(df['ESCALA_MOEDA'].value_counts())

    # Tipos de exercício contábil (se existir)
    if 'ORDEM_EXERC' in df.columns:
        print("\nOrdem do exercício:")
        print(df['ORDEM_EXERC'].value_counts())
```

## Principais Observações da EDA

A análise exploratória dos dados da CVM revelou alguns aspectos relevantes para o processamento posterior dos datasets:

- Os dados apresentam diferentes **escalas monetárias**, com parte das empresas reportando valores em milhares. Dessa forma, será necessário padronizar os valores financeiros para garantir comparabilidade entre as observações.

- Observou-se a presença de registros relacionados a diferentes **ordens de exercício contábil** (por exemplo, último e penúltimo exercício), o que pode gerar registros duplicados para uma mesma empresa e período. Será necessário definir um critério de filtragem na etapa de limpeza.

- Algumas empresas encontram-se em **recuperação judicial**, o que pode impactar significativamente indicadores financeiros e deve ser considerado na interpretação dos resultados.

- A análise das colunas `CD_CONTA` e `DS_CONTA` permitiu identificar as **contas contábeis relevantes** que serão utilizadas no cálculo dos indicadores financeiros na etapa de transformação dos dados.

Com base nessas observações, serão aplicadas regras de limpeza e transformação no pipeline ETL, incluindo padronização monetária, filtragem de exercícios contábeis e seleção de contas específicas para cálculo dos indicadores.
